```python
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import FunctionTransformer
from sklearn.impute import SimpleImputer

def Cumul_One_Hot(X, threshold=0.95):
    """
    One hot encode the columns where we are below the threshold.
    """

    X = X.copy()

    for col in categorical_cols:
        value_counts = X[col].value_counts(normalize=True)
        cumsum = value_counts.cumsum()

        categories = cumsum[cumsum < threshold].index

        X[col] = X[col].apply(lambda x: x if x in categories else 'Other')

    X = pd.get_dummies(X, columns=categorical_cols, drop_first=True)

    return X

column_transformer = ColumnTransformer(
    transformers=[
        ('impute', SimpleImputer(strategy='most_frequent'), X.columns),
        ('cumul_one_hot', FunctionTransformer(Cumul_One_Hot), X.columns)
    ]
)
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
import numpy as np
from sklearn.preprocessing import FunctionTransformer


def preprocess_data(X: pd.DataFrame) -> pd.DataFrame:

    X = X.copy()

    for col in X.columns:
        if X[col].nunique(dropna=True) == 2:
            unique_values = X[col].dropna().unique()
            mapping = {value: idx for idx, value in enumerate(unique_values)}
            X[col] = X[col].map(mapping)

    X['Profession'] = X['Profession'].fillna('Student')

    X['Pressure'] = X['Work Pressure'].fillna(0) + X['Academic Pressure'].fillna(0)

    X['Satisfaction'] = X['Job Satisfaction'].fillna(0) + X['Study Satisfaction'].fillna(0)

    np.random.seed(42)
    X['Sleep Duration'] = X['Sleep Duration'].apply(
        lambda value: (
            0 if value in ["Less than 5 hours", "2-3 hours", "3-4 hours", "4-5 hours", "No"]
            else 1 if value in ["5-6 hours", "6-7 hours", "4-6 hours"]
            else 2 if value in ["7-8 hours"]
            else 3 if value in ["8-9 hours", "More than 8 hours", "9-11 hours", "10-11 hours"]
            else np.random.choice([0, 1, 2, 3])
        )
    )

    X['Dietary Habits'] = X['Dietary Habits'].apply(
        lambda value: (
            0 if value in ["Unhealthy", "No Healthy", "No", "1", 1]
            else 1 if value in ["Moderate", "Less Healthy", "Less than Healthy", "2", 2]
            else 2 if value in ["Healthy", "More Healthy", "Yes", "3", 3]
            else np.random.choice([0, 1, 2])
        )
    )
    

    X = X.drop(['Work Pressure', 'Academic Pressure', 'Job Satisfaction', 'Study Satisfaction'], axis=1)

    return X


def replace_rare_categories(X, threshold=0.95):
    """
    Remplace les catégories rares par 'Other' pour les colonnes catégorielles
    selon un seuil cumulé de fréquence.
    """

    X = X.copy()
    categorical_cols = X.select_dtypes(include='object').columns

    for col in categorical_cols:
        value_counts = X[col].value_counts(normalize=True)
        cumsum = value_counts.cumsum()

        frequent_categories = cumsum[cumsum < threshold].index.tolist()

        X[col] = X[col].where(X[col].isin(frequent_categories), 'Other')
    
    return X

rare_category_transformer = FunctionTransformer(replace_rare_categories, validate=False)
    
class CGPACategorizer(BaseEstimator, TransformerMixin):
    """
    Categorize the CGPA column.
    """
    def __init__(self, q=4, columns=None):
        self.q = q
        self.bin_edges = None
        self.columns = columns

    def fit(self, X, y=None):
        if self.columns is None:
            self.columns = X.columns
        if 'CGPA' not in self.columns:
            raise ValueError("You must specify the 'CGPA' column explicitly.")

        self.bin_edges = pd.qcut(
            X['CGPA'].replace([np.inf, -np.inf], np.nan),
            q=self.q,
            duplicates='drop',
            retbins=True
        )[1]
        return self

    def transform(self, X):
        X = X[self.columns].copy()
        X['CGPA'] = pd.cut(
            X['CGPA'].replace([np.inf, -np.inf], np.nan),
            bins=self.bin_edges,
            labels=False,
            include_lowest=True
        ).fillna(self.q).astype(int)

        return X['CGPA']
    
```

```python
from xgboost import XGBClassifier
import optuna  #type: ignore
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.pipeline import Pipeline


def objective(trial):

    threshold = trial.suggest_float('preprocessing__one_hot__threshold', 0.5, 0.99)
    q_cgpa = trial.suggest_int('preprocessing__cgpa__q', 3, 6)

    params = {
        'n_estimators': trial.suggest_int('model__n_estimators', 100, 1000),
        'max_depth': trial.suggest_int('model__max_depth', 3, 10),
        'learning_rate': trial.suggest_float('model__learning_rate', 0.01, 0.3),
        'subsample': trial.suggest_float('model__subsample', 0.5, 1.0),
        'colsample_bytree': trial.suggest_float('model__colsample_bytree', 0.5, 1.0),
        'gamma': trial.suggest_float('model__gamma', 0, 5),
        'reg_alpha': trial.suggest_float('model__reg_alpha', 0, 10),
        'reg_lambda': trial.suggest_float('model__reg_lambda', 0, 10),
        'scale_pos_weight': trial.suggest_float('model__scale_pos_weight', 1, 10)
    }

    pipeline.set_params(
        preprocessing__one_hot__threshold=threshold,
        preprocessing__cgpa__q=q_cgpa,
        model__n_estimators=params['n_estimators'],
        model__max_depth=params['max_depth'],
        model__learning_rate=params['learning_rate'],
        model__subsample=params['subsample'],
        model__colsample_bytree=params['colsample_bytree'],
        model__gamma=params['gamma'],
        model__reg_alpha=params['reg_alpha'],
        model__reg_lambda=params['reg_lambda'],
        model__scale_pos_weight=params['scale_pos_weight']
    )

    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    scores = cross_val_score(pipeline, X_train, y_train, cv=cv, scoring='accuracy', n_jobs=-1)

    return np.mean(scores)
```
